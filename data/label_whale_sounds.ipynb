```python
import aifc
import numpy as np
import os
import pandas as pd
import json

base_dir = "/Users/janakagrawal/Downloads/whale-detection-challenge/data/"
```

```python
df = pd.read_csv("/Users/janakagrawal/Downloads/whale-detection-challenge/data/train.csv")
```

```python
def extract_peak_frequency(data, sampling_rate):
    fft_data = np.fft.fft(data)
    freqs = np.fft.fftfreq(len(data))
    
    peak_coefficient = np.argmax(np.abs(fft_data))
    peak_freq = freqs[peak_coefficient]
    
    return abs(peak_freq * sampling_rate)
```

```python
labels = {}
for index, row in df.iterrows():
    if row["label"] == 1:
        file_name = row["clip_name"]
        aud = aifc.open(os.path.join(base_dir, "train", file_name))
        b = aud.readframes(aud.getnframes())
        n = [x-256 if x>=128 else x for x in b]
        labels[file_name] = extract_peak_frequency(n, aud.getframerate())
```
