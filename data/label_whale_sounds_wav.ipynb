```python
import aifc
import numpy as np
import os
import pandas as pd
import json
from scipy.io import wavfile

base_dir = "/Users/janakagrawal/Downloads/3second"
```

```python
def extract_peak_frequency(data, sampling_rate):
    fft_data = np.fft.fft(data)
    freqs = np.fft.fftfreq(len(data))
    
    peak_coefficient = np.argmax(np.abs(fft_data))
    peak_freq = freqs[peak_coefficient]
    
    return abs(peak_freq * sampling_rate)
```

```python
labels = {}
for file_name in os.listdir(base_dir):
    try:
        samplerate, data = wavfile.read(os.path.join(base_dir, file_name))
        labels[file_name] = extract_peak_frequency(data, samplerate)
    except:
        print(file_name)
```
